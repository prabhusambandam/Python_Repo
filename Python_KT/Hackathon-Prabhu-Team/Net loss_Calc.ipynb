```python
## Taking Random Samples from the Record 
sample_means = []
n = 25 
for sample in range(0,300):
    #Random choice with replacement
    sample_values = np.random.choice(a = data['Net Loss (Barrels)'],size = n)
    sample_mean = np.mean(sample_values)
    sample_means.append(sample_mean)
```

```python
sns.distplot(sample_means)
plt.title("Distribution of the Sample Means $(n=25)$ records")
plt.xlabel("Net Loss (Barrels)")
```
