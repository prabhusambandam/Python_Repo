```python
order_data.head()
```

```python
sales_data.head()
```

```python
pd.merge(order_data,sales_data,on = 'Company Id')
```

```python
sales_data.rename(columns = {'Company Id':'Sales Order Id'},inplace = True)
```

```python
sales_data.head()
```
